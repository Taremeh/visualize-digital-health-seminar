# General Statistics Overview

## Import data

```python
import pandas as pd
import matplotlib.pyplot as plt

# define file paths
filepath = "ClinicalMarkers_final.csv"
# load data frame
df = pd.read_csv(filepath)

df
```

## General Statistics

```python
# Display general statistics
df.describe(include='all')
```

### Age statistics

```python
# Convert Age column to integers
df['Age'].describe()
```

```python
# Plot age distribution
plt.figure(figsize=(12, 4))
plt.hist(df['Age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Year of Birth')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Sex Statistics 

```python
# Sex statistics
sex_counts = df['sex'].value_counts()
sex_counts
```

```python
# Plot sex counts
plt.figure(figsize=(4, 4))
sex_counts.plot(kind='bar', color=['lightcoral', 'lightblue'])
plt.title('Sex Counts')
plt.xlabel('Sex')
plt.ylabel('Counts')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Disease Statistics

```python
# Disease_type statistics
disease_counts = df["disease_type"].value_counts()
disease_counts
```

```python
# Plot disease counts
plt.figure(figsize=(12, 4))
disease_counts.plot(kind='bar', color=['lightgreen', 'lightcoral', 'lightblue'])
plt.title('Disease Type Counts')
plt.xlabel('Disease Type')
plt.ylabel('Counts')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
